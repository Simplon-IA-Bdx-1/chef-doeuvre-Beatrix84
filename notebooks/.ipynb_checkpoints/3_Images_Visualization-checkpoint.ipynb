# Voxel coordinates are coordinates in the image data array
A voxel is a pixel with volume.

In the code above, slice_0 from the data is a 2D slice from a 3D image. The plot of the EPI slices displays the slices in grayscale (graded between black for the minimum value, white for the maximum). Each pixel in the slice grayscale image also represents a voxel, because this 2D image represents a slice from the 3D image with a certain thickness.

The 3D array is therefore also a voxel array. As for any array, we can select particular values by indexing. For example, we can get the value for the middle voxel in the EPI data array like this:

```python
n_i, n_j, n_k = im.shape
center_i = (n_i - 1) // 2  # // for integer division
center_j = (n_j - 1) // 2
center_k = (n_k - 1) // 2
center_i, center_j, center_k
```

The values (127, 127, 84) are indices into the data array epi_img_data. (127, 127, 84) is therefore a ‘voxel coordinate’ - a coordinate into the voxel array.
A coordinate is a set of numbers giving positions relative to a set of axes. In this case 127 is a position on the first array axis, where the axis is of length im.shape[0], and therefore goes from 0 to 255 (im.shape == (256, 256, 170)). 

```python
center_vox_value = im[center_i, center_j, center_k]
center_vox_value
```

# Images type

Images are a memory-map stored in a binary file on disk. Memory-mapped files are used for accessing small segments of large files on disk, without reading the entire file into memory. NumPy’s memmap’s are array-like objects.

# Visualize all slices for 1 Image

```python
try:
    from skimage.util.montage import montage2d
except ImportError as e:
    print('scikit-image is too new, ',e)
    from skimage.util import montage as montage2d

fig, ax1 = plt.subplots(1, 1, figsize = (20, 20))
ax1.imshow(montage2d(im), cmap ='bone')
fig.savefig('brain_scan.png')
```

Between the 70th and the 90th slices, grey matter si visible and comparable

```python
# Then we have a look at slices over the first, second and third dimensions of the array.
# https://nipy.org/nibabel/coordinate_systems.html
from scipy import ndimage

def show_slices(slices):
#Function to display row of image slices
    fig, axes = plt.subplots(1, len(slices), figsize = (12, 6))
    for i, slice in enumerate(slices):

        axes[i].imshow(slice.T, cmap="gray", origin="lower")
        
im = ndimage.rotate(im, 180, reshape=False)
slice_0 = im[83, :, :]
slice_1 = im[:, 150, :]
slice_2 = im[:, :, 83]


show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Center slices for Brain image")  
```
